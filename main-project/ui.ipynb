```python
import tkinter as tk
from tkinter import messagebox
import database

database.create_table()

def add():
    database.add_student(name_entry.get(), marks_entry.get())
    messagebox.showinfo("Success", "Student Added")

def view():
    data = database.get_students()
    output.delete("1.0", tk.END)
    for row in data:
        output.insert(tk.END, str(row) + "\n")

def search():
    data = database.search_student(name_entry.get())
    output.delete("1.0", tk.END)
    for row in data:
        output.insert(tk.END, str(row) + "\n")

def update():
    database.update_student(id_entry.get(), marks_entry.get())
    messagebox.showinfo("Success", "Updated")

def delete():
    database.delete_student(id_entry.get())
    messagebox.showinfo("Success", "Deleted")

root = tk.Tk()
root.title("Student Management System")

tk.Label(root, text="ID").pack()
id_entry = tk.Entry(root)
id_entry.pack()

tk.Label(root, text="Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Marks").pack()
marks_entry = tk.Entry(root)
marks_entry.pack()

tk.Button(root, text="Add", command=add).pack()
tk.Button(root, text="View", command=view).pack()
tk.Button(root, text="Search", command=search).pack()
tk.Button(root, text="Update", command=update).pack()
tk.Button(root, text="Delete", command=delete).pack()

output = tk.Text(root, height=10)
output.pack()

root.mainloop()
```
